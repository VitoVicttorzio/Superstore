# Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import matplotlib.dates as mpl_dates
%matplotlib inline
import seaborn as sns 
import plotly.graph_objects as go
```

# Reading dataset

```python
df = pd.read_excel(r"Sales_Superstore.xlsx", header=0, index_col=0)
df.index.name = None
```

```python
# Display first 10 rows of dataframe

pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:.2f}".format)
df.head(10)
```

# Depict the distributions by using Box Plots

```python
# Plotting boxplot for Quantity

plt.figure(figsize=(10, 3))
sns.boxplot(df["Quantity"], orient="h", width=0.5, 
            linewidth=3, linecolor="black", fliersize=7)
plt.xlabel("Quantity", size=15, labelpad=15.0)
plt.show()
```

```python
# Creating boxplot for Discount

plt.figure(figsize=(10,3))
sns.boxplot(df['Discount'], orient='h', color='plum', width=0.5,
            linewidth=3, linecolor='black', fliersize=7)
plt.xlabel("Discount", size=15, labelpad=15.0)
plt.show()
```

# Correlation between variables

```python
# Building a correlation heatmap

plt.figure(figsize=(7,5))
correlation = df[['Sales', 'Quantity', 'Discount', 'Profit']]
sns.heatmap(correlation.corr(), cmap= 'coolwarm',  fmt=".2f", annot=True,  linewidth=3.,
            vmin=-1, vmax=1, annot_kws={'fontsize': 12, 'fontweight': 'bold'}, square=True)
plt.show()
```

```python
# Building scatter plot

plt.figure(figsize=(8,6))
sns.regplot(x='Sales', y='Profit', data=df, ci=None, scatter_kws={'s':100, 'alpha' : 0.75, 'edgecolor': 'black'})
plt.title('Correlation between Sales and Profit', size=20, fontweight='bold')
plt.xlabel('Sales', size=15, labelpad=15.0)
plt.ylabel('Profit', size=15, labelpad=15.0)
plt.xticks(np.arange(0, max(df['Sales']) + 1, 2000))
plt.yticks(np.arange(-7000, max(df['Profit']) + 1, 1500))
plt.show()
```

# Analysing dataframe by segments

```python
# Count unique segments

segments = df['Segment'].nunique()
print(f'There are {segments} segments in the dataframe')
```

```python
# Count customers by Segment

number_of_customers = df['Segment'].value_counts().reset_index()
number_of_customers = number_of_customers.rename(columns={'count' :'Total Customers'})
number_of_customers
```

```python
# Plotting Pie chart

plt.figure(figsize=(10,5))
plt.pie(number_of_customers['Total Customers'], labels=number_of_customers['Segment'], autopct = '%1.1f%%',
        pctdistance=0.5, labeldistance=1.1, startangle=90, shadow=True)
plt.title('Distributition of customers by segment', weight='bold', fontsize=15)
plt.legend(loc='center left', borderaxespad=30)
plt.show()
```

```python
# Groupping sales and profit per category

sales_per_category = df.groupby('Segment')['Sales'].sum().reset_index()
profit_per_category = df.groupby('Segment')['Profit'].sum().reset_index()
```

```python
# Plotting Bar chart based on grouping sales and profit

fig, ax = plt.subplots(figsize=(10, 6)) 
ax.bar(sales_per_category['Segment'], sales_per_category['Sales'])
ax.bar(profit_per_category['Segment'], profit_per_category['Profit'])

for s in ['top', 'right']:
    ax.spines[s].set_visible(False)

for x, y in enumerate(sales_per_category['Sales']):
    plt.text(x, y, str(round(y)), ha='center', va='bottom', size=12)

for x, y in enumerate(profit_per_category['Profit']):
    plt.text(x, y, str(round(y)), ha='center', va='bottom', size=12)
    
plt.title("Sales and profit by segment", weight='bold', size=20, pad=30.0)
plt.xlabel("Segment", size=15, labelpad=15.0)
plt.ylabel("Sales and Profit", size=15, labelpad=15.0)
plt.ticklabel_format(style='plain', axis='y')
plt.yticks(np.arange(0, max(sales_per_category['Sales']) + 1, 100000))

plt.legend(['Sales', 'Profit'], loc = 'upper right')
plt.show()
```

```python
# Plotting horisontal bar chart 'Sales by sub-category'

top_subcategory = product_subcategory.sort_values(by='Sales', ascending=True)
fig, ax = plt.subplots(figsize=(10, 7)) 

colors = ['r' if el < 100_000 else ('b' if el < 200_000 else 'g') for el in top_subcategory['Sales']]
ax.barh(top_subcategory['Sub-Category'], top_subcategory['Sales'], color=colors)

for s in ['top', 'right']:
    ax.spines[s].set_visible(False)

ax.grid(axis='x', color='grey',linestyle='-', alpha=0.1)    

for x, y in enumerate(top_subcategory['Sales']):
    ax.text(y, x, str(round(y)), va='center', size=10, fontweight='bold')

plt.title("Sales by Sub-Category", weight='bold', size=20)
plt.ylabel("Sub-category", size=15, labelpad=15.0)
plt.xlabel("Total Sales", size=15, labelpad=15.0)
plt.show()
```

```python
# Groupping data by Sub-Category to calculate profit

product_subcategory = df.groupby('Sub-Category')['Profit'].sum().reset_index()
top_subcategory = product_subcategory.sort_values(by='Profit', ascending=False).reset_index(drop=True)
top_subcategory.head()
```

```python
# Plotting Bar chart 'Profit by sub-category'

top_subcategory = top_subcategory.sort_values(by='Profit', ascending=True).reset_index(drop=True)
fig, ax = plt.subplots(figsize=(10, 7)) 
colors = ['g' if el > 15_000 else ('b' if el > 0  else 'r') for el in top_subcategory['Profit']]
ax.bar(top_subcategory['Sub-Category'], top_subcategory['Profit'], color=colors)

for s in ['top', 'right']:
    ax.spines[s].set_visible(False)

ax.grid(axis='y', color='grey',linestyle='-', alpha=0.2)    

for x, y in enumerate(top_subcategory['Profit']):
    if y > 0:
        ax.text(x, y, str(round(y)), ha='center', va='bottom', size=10, fontweight='bold')
    elif y < 0:
        ax.text(x, y, str(round(y)), ha='center', va='top', size=10, fontweight='bold')
    
plt.title("Profit by Sub-Category", weight='bold', size=18)
plt.ylabel("Total Profit", size=15, labelpad=15.0)
plt.xlabel("Sub-category", size=15, labelpad=15.0)
plt.xticks(rotation=50)
plt.show()
```

# Sales Trend Analysis

```python
# Divide data by month

year_sales = df.loc[:,['Order Date', 'Sales', 'Profit']]
year_sales = year_sales.resample('ME', on='Order Date')[['Profit', 'Sales']].sum().reset_index()
year_sales.head()
```

```python
# Create two line graph based on Sales and Profit by month

plt.figure(figsize=(10,12))

plot_1 = plt.subplot(2, 1, 1)
plot_1.plot(year_sales['Order Date'], year_sales['Sales'], marker='.', markersize=15, linewidth=3)
date_format = mpl_dates.DateFormatter('%m-%Y')
plt.gca().xaxis.set_major_formatter(date_format)
plt.grid(axis='y', color='grey',linestyle='-', alpha=0.3)   

for s in ['top', 'right']:
    plot_1.spines[s].set_visible(False)
    
plt.title("Dynamics of Sales by month", weight='bold', size=20)
plt.yticks(np.arange(0, max(year_sales['Sales']) + 1, 10000))
plt.xticks(rotation=30)
plt.xlabel('Date', size=15, labelpad=15.0)
plt.ylabel('Sales', size=15, labelpad=15.0)

plot_2 = plt.subplot(2, 1, 2)
plot_2.plot(year_sales['Order Date'], year_sales['Profit'], marker='.', markersize=15, linewidth=3, color='g')
date_format = mpl_dates.DateFormatter('%m-%Y')
plt.gca().xaxis.set_major_formatter(date_format)

for s in ['top', 'right']:
    plot_2.spines[s].set_visible(False)
    
plt.title("Dynamics of Profit by month", weight='bold', size=20)
plt.yticks(np.arange(-5000, max(year_sales['Profit']) + 1, 2500))
plt.xticks(rotation=30)
plt.grid(axis='y', color='grey',linestyle='-', alpha=0.35)    
plt.xlabel('Date', size=15, labelpad= 5.0)
plt.ylabel('Profit', size=15, labelpad=15.0)

plt.subplots_adjust(hspace=0.5)
plt.show()
```
